# Data Visualization

```python
# Top 5 companies by Percentage_laid_off
company_percentage_layoff = layoff.groupby('company')['percentage_laid_off'].sum().reset_index()
top_5_company = company_percentage_layoff.sort_values(by='percentage_laid_off', ascending=False).head(5)
top_5_company.set_index('company')['percentage_laid_off'].plot(kind='pie', autopct='%1.1f%%', shadow = True, explode = (0,0,0,0,0.1))
plt.title('Top 5 Companies by Percentage of Layoffs')
plt.show()
```

```python
#Top 10 countries by total_laid_off
countries_total_layoff = layoff.groupby('country')['total_laid_off'].sum()
top_10_countries = countries_total_layoff.sort_values(ascending=False).head(10)
top_10_countries.plot(kind='bar')
plt.title('Top 10 Countries by Total Layoffs')
plt.ylabel('total_laid_off')
plt.xlabel('country')
plt.show()
```

```python
# Total numbers of each industry
layoff.industry.value_counts()
```

```python
# Display the industry 
layoff['industry'].value_counts().plot(kind='bar',figsize=(10,6))
plt.xlabel('Industry')
plt.ylabel('Total_laid_off')
plt.title('Total laid off per industry')
plt.legend()
```

```python
#top 10 countries by fund_raised_million
countries_fund_raised = layoff.groupby('country')['funds_raised_millions'].sum()
top_10_countries = countries_fund_raised.sort_values(ascending=False).head(10)
top_10_countries.plot(kind='bar' , color = 'green')
plt.title('Top 10 Countries by fund_raised_million')
plt.ylabel('fund_raised_million')
plt.xlabel('country')
plt.show()
```
